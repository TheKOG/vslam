```python
import stereo_depth_online
import matplotlib.pyplot as plt
stereo_depth_online.Init("stereoDepth/stereo_cam.yml")
```

```python
import cv2
origin=cv2.imread("stereoDepth/origin_images/cali2.png")
# origin=cv2.imread("origin.png")
height,width,c=origin.shape
left = origin[0:height, 0:int(width/2), :]
right = origin[0:height, int(width/2):width, :]

image_rgb = cv2.cvtColor(origin, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
```

```python
dis=stereo_depth_online.depth_map(left,right)

image_rgb = cv2.cvtColor(dis, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
```

```python
dis=stereo_depth_online.stereo_depth(left,right)

image_rgb = cv2.cvtColor(dis, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
```
